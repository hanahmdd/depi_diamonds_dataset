```python
import pandas as pd

diamonds = pd.read_csv("/content/diamonds.csv")
diamonds.head()
```

```python
diamonds.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
diamonds["cut"].value_counts()
```

```python
diamonds["cut"] = diamonds["cut"].str.title()
```

```python
diamonds["cut"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.violinplot(x="cut", y="price", data=diamonds, inner="quartile")
plt.title("Diamond Price Distribution by Cut")
plt.show()
```

```python
diamonds["color"].value_counts()
```

```python
diamonds = diamonds[diamonds["color"].isin(["D","E","F","G","H","I","J"])]
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="color", y="price", data=diamonds)
plt.title("Diamond Price by Color")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="carat", y="price", data=diamonds, alpha=0.4)
plt.title("Carat vs Price")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="carat", y="price", hue="cut", data=diamonds, alpha=0.5)
plt.title("Carat vs Price by Cut")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(diamonds.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
